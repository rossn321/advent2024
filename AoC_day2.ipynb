```python
safe_reports = 0
reports = []

# turn the reports into lists of ints
for line in lines:
  reports.append(list(map(int, line.split())))

# check safety of reports
for report in reports:
  ascending = None
  for i in range(len(report) - 1):
    # exit the loop if two numbers are equal, unsafe
    if report[i] == report[i + 1]:
      break
    # exit the loop if numbers have an absolute difference greater than 3, unsafe
    elif abs(report[i + 1] - report[i]) > 3:
      break
    # ascending/descending check: we want to consistently do one and not swap between
    # check if ascending, exit the loop if we are descending, unsafe
    elif report[i + 1] > report[i]:
      if ascending is None:
        ascending = True
      elif not ascending:
        break
    # check if descending, exit the loop if we are ascending, unsafe
    elif report[i + 1] < report[i]:
      if ascending is None:
        ascending = False
      elif ascending:
        break
  # if we make it this far, the report is safe
  else:
    safe_reports += 1

print(safe_reports)
```

```python
# Part 2 Solution


# turn the reports into lists of ints
reports = []
for line in lines:
  reports.append(list(map(int, line.split())))

# function to check if report is safe under the original part 1 constraints
def is_safe(report):
  '''
  Checks if a report is safe based on the original constraints:
  - Monotonically increasing or decreasing
  - Adjacent elements differ by at most 3
  '''
  ascending = None # initialise to track ascending or decreasing trends

  # Check for unsafe differences between elements (0 or >3)
  for i in range(len(report) - 1):
    difference = report[i + 1] - report[i]
    if difference ==0 or abs(difference) > 3:
      return False
    # ascending/descending check: we want to consistently do one and not swap between
    if ascending is None:
      ascending = (difference > 0)
    elif (difference > 0) != ascending:
      return False

  # if we make it this far, the report is safe
  return True

def count_safe_reports_with_dampener(reports):
  """
  Counts the number of safe reports, considering the Problem Dampener.
  A report is safe if it's originally safe OR if removing a single
  element makes it safe.
  """
  safe_count = 0
  for report in reports:
    if is_safe(report) == True: # check if already safe without dampener, like in part 1
      safe_count += 1
    else:
      for i in range(len(report)): # iterate through report, removing an element at a time to see if safe when one element removed
        temp_report = report[:i] + report[i+1:]
        if is_safe(temp_report) == True:
          safe_count += 1
          break # exit the loop if we find a safe configuration
  return safe_count

print(count_safe_reports_with_dampener(reports))
```
